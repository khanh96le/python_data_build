# Transform

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
index = pd.date_range('10/1/1999', periods=1100)
```

```python
index
```

```python
ts = pd.Series(np.random.normal(0.5, 2, 1100),index)
ts
```

If you are wondering what is **numpy.normal** means? This is the answer

```python
plt.plot(np.random.normal(0.5, 2, 1000))
```

```python
plt.plot(np.random.normal(10, 20, 1000))
```

```python
ts.rolling(window=100, min_periods=100).mean().dropna()
```

```python
key = lambda x: x.year
```

```python
zscore = lambda x: (x - x.mean()) / x.std()
```

```python
transformed = ts.groupby(key).transform(zscore)
transformed
```
