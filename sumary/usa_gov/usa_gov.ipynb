> URL shortening service bit.ly partnered with the United States governmentwebsite usa.gov to provide a feed of anonymous data gathered from users who shortenlinks ending with .gov or .mil

> In that data, each record contain infomation about timezone of user who shortenlinks.

## Question 1: What is the top 10 city where people in there use bit.ly most?

To answer this question, we need to get the place data in our dataset. Then count them and rank them.

```python
import json
path_usa = "data/usa_gov_data.txt"
records = [json.loads(line) for line in open(path_usa)]
```

```python
records
```

```python
records = [json.loads(line) for line in open(path_usa)]
timezones = [rec['tz'] for rec in records if 'tz' in rec]
timezones
```

```python
def get_counts(sequence):
  counts = {}
  for x in sequence:
    if (x in counts):
      counts[x] = counts[x] + 1
    else:
      counts[x] = 1
  return counts
counts = get_counts(timezones)
counts
```

```python
from pandas import DataFrame, Series
import pandas as pd
frame = DataFrame(records)
frame
```

```python
tz_counts = frame['tz'].value_counts()
tz_counts
```

Now we need to clean and mung data

```python
clean_tz = frame['tz'].fillna('Missing')
clean_tz[clean_tz == ''] = 'Unknown'
clean_tz
```

Then visualize top 10 places where use .bitly most

```python
import matplotlib.pyplot as plt
tz_counts[:10].plot(kind='barh', rot=0)
plt.show()
```

## Question 2: What is the percent of Windows OS people use in each city?

To answer this question, we need to do 2 tasks:
- Get all data contain info about the OS.
- There are a lot of OS, but we care only about Window and Not Window. So we need to transform the messy data into a clear format contain only 2 state: Window and Not Window.

```python
frame
```

```python
cframe = frame[frame.a.notnull()]
import numpy as np
os = np.where(cframe["a"].str.contains("Windows"), "Windows", "Not Windows")
os
```

```python
by_tz_os = cframe.groupby(['tz', os])
by_tz_os.size()
agg_counts = by_tz_os.size().unstack()
agg_counts
```

```python
agg_counts.fillna(0)
```

```python
indexer = agg_counts.sum(1).argsort()
```

```python
count_subset = agg_counts.take(indexer)[-10:]
count_subset.plot(kind='barh', stacked='True', rot=0)
plt.show()
```
